```python
#Importing the libraries
import json 
import time 

sales = ' '

#Loading the records from JSON to string 
fd = open("Records.json","r")
js = fd.read()
fd.close()

#Converting String records to JSON/ Dict 
record = json.loads(js)

#Displaying Menu
print("----------------------MENU----------------------\n")

for key in record.keys():
    print(key,":", record[key]["Name"],"\t|",record[key]["Price"],"\t|",record[key]["Qn"])
    
print("------------------------------------------------\n")

#Taking user inputs about the purchase and the user details 
ui_name = str(input("Enter your name     :"))
ui_mail = str(input("Enter your mail     :"))
ui_ph   = str(input("Enter your phone no :"))
ui_pr   = str(input("Enter Product ID    :"))
ui_qn   = int(input("Enter Quantity      :"))

#if we're having more or enough quantity 

if(record[ui_pr]["Qn"]>= ui_qn):
    
    #Displaying the purchase details 
    print("------------------------------------------------\n")
    print("                      BILL                      \n")
    print("Name           :",record[ui_pr]["Name"])
    print("Price          :",record[ui_pr]['Price'],"Rs")
    print("Quantity       :",ui_qn)
    
    
    print("------------------------------------------------")
    print("Billing amount :", record[ui_pr]["Price"]*ui_qn,"Rs")
    print("------------------------------------------------")
    
    #Generating Sales 
    sales = ui_name + "," +ui_mail + "," + ui_ph + "," + ui_pr + "," +record[ui_pr]["Name"] + "," +str(ui_qn) + "," +str(record[ui_pr]["Price"]*ui_qn)+','+time.ctime()+'\n'
    
    #updating records dictionary 
    record[ui_pr]["Qn"] = record[ui_pr]["Qn"] - ui_qn

#if we're not having enough quantity    
else:
    #Displaying info about the quantity 
    print("Sorry we are not having enough quantity in our Inventory")
    print("We're only having",record[ui_pr]['Qn'],"quantity.")
    print("------------------------------------------------")
    
    #if user wants to purchase whole quantity for that product 
    ch = str(input("Press Y/y to purchase:"))
    if(ch =="Y" or ch =="y"):
        
        #Displaying the purchase details
        print("------------------------------------------------\n")
        print("                      BILL                      \n")
        print("Name           :",record[ui_pr]["Name"])
        print("Price          :",record[ui_pr]['Price'],"Rs")
        print("Quantity       :",record[ui_pr]["Qn"])
        print("------------------------------------------------")
        print("Billing amount :", record[ui_pr]["Price"]*record[ui_pr]["Qn"],"Rs")
        print("------------------------------------------------")
        
        #Generating Sales       
        sales = ui_name + "," +ui_mail + "," + ui_ph + "," + ui_pr + "," +record[ui_pr]["Name"] + "," +str(ui_qn) + "," +str(record[ui_pr]["Price"]*record[ui_pr]["Qn"])+','+time.ctime()+'\n'
        
        #updating records dictionary
        record[ui_pr]['Qn'] = 0
    else:
        print("Thanks")
    



#Saving records in JSON file 
js = json.dumps(record)

fd = open("Records.json","w")
fd.write(js)
fd.close()

# saving records in Sales file 
fd = open("Sales.txt","a")
fd.write(sales)
fd.close()

print("----------Thanks for purchasing--------")
```
